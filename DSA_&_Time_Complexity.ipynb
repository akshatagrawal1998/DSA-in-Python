```python
n = int(input("Enter n: "))

for i in range(2**n):
    print(i)


# time complexity is: 2 raise to the power n
```

```python
# recursive fibonacci series code:
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

n = int(input("Enter n: "))
print(fibonacci(n))

# exponential time complexity (2 raise to the power n)
# har step me 2 steps bdh ja rhe hain

```
