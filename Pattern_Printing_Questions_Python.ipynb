# Pattern 27:
Screenshot 2026-02-24 at 9.34.05 AM.png


```python
n = 5

for i in range(1, n+1):
  print(" "*(n-i), end = " ")
  c = 64
  for j in range((2*i)-1):
    if j < (i):
      c = c + 1
      print(chr(c), end = "")
    else:
      c = c - 1
      print(chr(c), end = "")
  print()
```

```python
# alternative solution

n = 5

for i in range(1, n + 1):

    # Print spaces
    print(" " * (n - i), end="")

    # Increasing letters
    for j in range(i):
        print(chr(65 + j), end="")

    # Decreasing letters
    for j in range(i - 2, -1, -1):
        print(chr(65 + j), end="")

    print()
```

```python
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


# print(is_prime(15))


num = 2

for i in range(1, 6):          # rows
    for j in range(1, i+1):    # numbers in each row

        for k in range(num, 100):   # search next prime
            if is_prime(k):
                print(k, end=" ")
                num = k + 1
                break

    print()
```

```python
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


num = 2

for i in range(1, 6):
    count = 0

    while count < i:
        if is_prime(num):
            print(num, end=" ")
            count += 1
        num += 1

    print()
```
